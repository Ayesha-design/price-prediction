```python
isb_urls = ["https://www.zameen.com/Houses_Property/Islamabad-3-1.html","https://www.zameen.com/Flats_Apartments/Islamabad-3-1.html","https://www.zameen.com/Penthouse/Islamabad-3-1.html","https://www.zameen.com/Farm_Houses/Islamabad-3-1.html","https://www.zameen.com/Rooms/Islamabad-3-1.html"]
lhr_urls =["https://www.zameen.com/Houses_Property/Lahore-1-1.html","https://www.zameen.com/Flats_Apartments/Lahore-1-1.html","https://www.zameen.com/Penthouse/Lahore-1-1.html","https://www.zameen.com/Farm_Houses/Lahore-1-1.html","https://www.zameen.com/Rooms/Lahore-1-1.html"]
karachi_urls=["https://www.zameen.com/Houses_Property/Karachi-2-1.html","https://www.zameen.com/Flats_Apartments/Karachi-2-1.html","https://www.zameen.com/Penthouse/Karachi-2-1.html","https://www.zameen.com/Farm_Houses/Karachi-2-1.html","https://www.zameen.com/Rooms/Karachi-2-1.html"]
peshawar_urls=["https://www.zameen.com/Houses_Property/Peshawar-17-1.html","https://www.zameen.com/Flats_Apartments/Peshawar-17-1.html","https://www.zameen.com/Penthouse/Peshawar-17-1.html","https://www.zameen.com/Farm_Houses/Peshawar-17-1.html","https://www.zameen.com/Rooms/Peshawar-17-1.html"]
rwd_urls=["https://www.zameen.com/Houses_Property/Rawalpindi-41-1.html","https://www.zameen.com/Flats_Apartments/Rawalpindi-41-1.html","https://www.zameen.com/Penthouse/Rawalpindi-41-1.html","https://www.zameen.com/Farm_Houses/Rawalpindi-41-1.html","https://www.zameen.com/Rooms/Rawalpindi-41-1.html"]
abbottabad_urls=["https://www.zameen.com/Houses_Property/Abbottabad-385-1.html","https://www.zameen.com/Flats_Apartments/Abbottabad-385-1.html","https://www.zameen.com/Penthouse/Abbottabad-385-1.html","https://www.zameen.com/Farm_Houses/Abbottabad-385-1.html","https://www.zameen.com/Rooms/Abbottabad-385-1.html"]
```

```python

property_type = ['House','Flat','penthouse','farmhouse','Room']
cities = ['Islamabad', 'Lahore','Karachi','Peshawar','Rawalpindi','Abbottabad']
```

```python
csv_file="zameen.csv"
```

```python
for city in cities:
    if city=="Islamabad":
        with open(csv_file, 'w', newline='') as csvfile:
          writer = csv.writer(csvfile)
          writer.writerow(['city', 'property_type','page_url', 'location', 'price', 'area', 'beds' , 'baths' ])
        data_collection("Islamabad",isb_urls)
    elif city=="Lahore":
        data_collection("Lahore",lhr_urls)
    elif city=="Karachi":
        data_collection("Karachi",karachi_urls)
    elif city=="Rawalpindi":
        data_collection("Rawalpindi",rwd_urls)
    elif city=="Peshawar":
        data_collection("Peshawar",peshawar_urls)
    else: 
        data_collection("Abbottabad",abbottabad_urls)
 
        
    
```

```python
def data_collection(city,urls):
    count=0
#    write()
    
    
    for url in urls: 
      response = requests.get(url)
      html = response.content

      soup = BeautifulSoup(html, 'html.parser')


      pages = [] #make an array for the pages 
    
      
      next_link = soup.find("a", title="Next")
      if next_link is not None:
        next_link = soup.find("a", title="Next")["href"]    
        pages.append(next_link)
      else:
        pages.append(url)

      while next_link is not None:
    # Make a request to the next page and parse the HTML content
       
        response = requests.get(url + next_link)
        soup = BeautifulSoup(response.text, "html.parser")

    # Find the link to the next page and add it to the list of pages
        next_link = soup.find("a", title="Next")
        if next_link is not None:
            next_link = next_link["href"]
            pages.append(next_link)
        
      property=property_type[count]
      print(len(pages))  
      
# Scrape the data from each page
      for page in range(0,len(pages)):
        
        response = requests.get(url + pages[page])
        soup = BeautifulSoup(response.text, "html.parser")
      
        data=[]
    
        locations_arr=[]
        prices_arr=[]
        areas_arr=[]
        beds_arr=[]
        baths_arr=[]
        property_url_arr=[]
    
          
      
        location = soup.find_all("div",  {'aria-label':'Location'} )
        for l in location:
            
          locations_arr.append(l.text)
            
        price = soup.find_all("span", {'class':'f343d9ce'})
        for p in price:
          prices_arr.append(p.text)
        area = soup.find_all("span",  {'aria-label':'Area'})
        for a in area:
          areas_arr.append(a.text)
        beds = soup.find_all("span",  {'aria-label':'Beds'})
        for b in beds:
          beds_arr.append(b.text)
        baths = soup.find_all("span",  {'aria-label':'Baths'})
        for ba in baths:
          baths_arr.append(ba.text)
        property_urls=soup.find_all('a', {'class':'_7ac32433'})   
        for a in property_urls:
                
          property_url_arr.append("zameen.com"+a["href"])    
    #    data.clear()      
          
        
    #    print(data)
        data.append({'property_url':property_url_arr,'location':locations_arr,'price':prices_arr,'area':areas_arr,'beds':beds_arr,'baths':baths_arr})  
        with open(csv_file, 'a', newline='') as csvfile:
            
            
        
        # Create a CSV writer
          writer = csv.writer(csvfile)            
         
          for item in data:
        # Write a row for each item
            for i in range(0,len(locations_arr)):
              try:
                  writer.writerow([city, property, property_url_arr[i], locations_arr[i], prices_arr[i], areas_arr[i], beds_arr[i],baths_arr[i]])  
              except IndexError:
                  print("")
      print("done with " + property)
      count=count+1
    
```
